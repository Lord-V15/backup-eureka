```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
```

```python
dataset=load_boston()
print(dataset.keys())
```

```python
print(dataset.DESCR)
```

```python
print(dataset.feature_names)
```

```python
print(dataset.data.shape)
```

```python
z=10
#WHICH FEATURE?
x=dataset.data[:,z]
y=dataset.target
plt.xlabel(dataset.feature_names[z])
plt.ylabel('Median Value')
plt.title("Dataset Features")
print(x.shape, y.shape)
plt.plot(x,y,'.')
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(dataset['data'],dataset['target'])
print("Shapes are: ",x_train.shape,y_train.shape,x_test.shape,y_test.shape)
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train, y_train)

print("At x=0, y= ",reg.intercept_,"\n")

plt.plot(np.arange(0,127),reg.predict(x_test),'r',label='Predictions')
# plt.plot(np.arange(0,127),y_test,'b',label='Actual Data')
plt.legend()
plt.show()
```

```python
plt.plot(reg.predict(x_test),y_test,'.',markersize=10)
plt.xlabel('Predicted Values')
plt.ylabel('True Values')
plt.title('Almost Linear Relatinon b/w predictions and testing data')
plt.show()
```

```python
print("MEAN SQUARED ERROR IS: ",sklearn.metrics.mean_squared_error(y_test, reg.predict(x_test)) )
```
