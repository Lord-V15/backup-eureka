```
import cv2
```

```
horse_actual = cv2.imread("horse.jpeg")
horse = cv2.cvtColor(horse_actual,cv2.COLOR_RGB2BGR)
horse_b,horse_g,horse_r = cv2.split(horse)

rainbow_actual = cv2.imread("rainbow.jpg")
rainbow = cv2.cvtColor(rainbow_actual,cv2.COLOR_RGB2BGR)
rainbow_b, rainbow_g, rainbow_r = cv2.split(rainbow)
```
