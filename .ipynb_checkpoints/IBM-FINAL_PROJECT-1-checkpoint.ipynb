```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import seaborn
import folium
```

```python
df = pd.read_csv('https://cocl.us/datascience_survey_data')
df.head()
```

```python
df=df.rename(columns={"Unnamed: 0": " "})
df.head(3)
```

```python
df.set_index(" ",inplace=True)
```

```python
df
```

```python
df=df.sort_values(by=['Very interested'],ascending=False)
df.head()
```

### TOTAL RESPONDEnTS = 2233

```python
df1=df.astype(float)
for i in range(6): 
    df1['Very interested'][i] = (df1['Very interested'][i]) / 22.33
    df1['Somewhat interested'][i] = (df1['Somewhat interested'][i]) / 22.33
    df1['Not interested'][i] = (df1['Not interested'][i]) / 22.33
(df1)
```

```python
for i in range(6):
    df1['Very interested'][i] = np.around(df1['Very interested'][i], decimals=2)
    df1['Somewhat interested'][i] = np.around(df1['Somewhat interested'][i], decimals=2)
    df1['Not interested'][i] = np.around(df1['Not interested'][i], decimals=2)
df1
```

```python
plt.figure(figsize=(20, 8))
val=np.array([3,6,9,12,15,18])
ax = plt.subplot(111)
ax.set_title("Percentage of Respondents' interest in Data Science Areas", fontdict={"size":16})

b1=ax.bar(x=val-1,height=df1['Very interested'],width=0.7,align='edge',color='#5cb85c',label='Very Interested')
b2=ax.bar(x=val,height=df1['Somewhat interested'],width=0.7,align='center',color='#5bc0de',label='Somewhat Interested')
b3=ax.bar(x=val+0.7,height=df1['Not interested'],width=0.7,align='center',color='#d9534f',label='Not Interested')

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(b1)
autolabel(b2)
autolabel(b3)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)

plt.xticks(val, df1.index.values, rotation='vertical')
plt.legend(prop={"size":14})
plt.show()
```
