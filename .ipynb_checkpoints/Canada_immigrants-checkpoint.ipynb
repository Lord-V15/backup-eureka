```python
india_arr=[]
years=[]
for i in range(1980,2013,+1):
    india_arr.append(df.loc['India',i])
    years.append(i)
```

```python
print(india_arr)
```

```python
print(years)
```

```python
fig = plt.figure(figsize=[12,5])
ax = plt.subplot(111)

#  remove tick marks
ax.xaxis.set_tick_params(size=0)
ax.yaxis.set_tick_params(size=0)

ax.set_title("Indian Immigrants in Canada b/w 1998-2013")
ax.set_xlabel('YEARS')
ax.set_ylabel('POPULATION')

# tweak the axis labels
xlab = ax.xaxis.get_label()
ylab = ax.yaxis.get_label()

xlab.set_style('italic')
xlab.set_size(15)
ylab.set_style('italic')
ylab.set_size(15)

# tweak the title
ttl = ax.title
ttl.set_weight('bold')
ttl.set_size(20)


plt.plot(years,india_arr,'--',color='green',label="No. of immigrants registered")
plt.legend()
plt.show()
```

## APPLYING MACHINE LEARNING TO LEARN PATTERN

```python
from sklearn.linear_model import LinearRegression
X = np.asarray(years)
X=X.reshape(-1,1)
y = np.asarray(india_arr)
```

```python
reg = LinearRegression().fit(X, y)
```

```python
reg.score(X, y)
```

```python
reg.predict(np.array([[2019]]))
```

```python
reg.predict(np.array([[2019]])) [0]
```

## MAKING NEW PLOT WITH PREDICTIONS

```python
india_new=[]
years_new=[]
```

```python
for i in range(1980,2023,+1):
    india_new.append( reg.predict(np.array([[i]]))[0] )
    years_new.append(i)
```

```python
print(india_new,"\n")
print(years_new)
```

```python
plt.plot(years_new,india_new,'--',c='b')
plt.plot(years,india_arr,'*',color='g')

plt.show()
```

```python
fig = plt.figure(figsize=[12,5])
ax = plt.subplot(111)

ax.set_title("Data of Indian Immigrants in Canada")
ax.set_xlabel('YEARS')
ax.set_ylabel('POPULATION')


l = ax.fill_between(years, india_arr, label="Given Data by IBM")
l.set_facecolors([[.5,.5,.8,.3]])

# change the edge color (bluish and transparentish) and thickness
l.set_edgecolors([[0, 0, .5, .3]])
l.set_linewidths([3])

#  remove tick marks
ax.xaxis.set_tick_params(size=0)
ax.yaxis.set_tick_params(size=0)

# change the color of the top and right spines to opaque gray
ax.spines['right'].set_color((.8,.8,.8))
ax.spines['top'].set_color((.8,.8,.8))

# tweak the axis labels
xlab = ax.xaxis.get_label()
ylab = ax.yaxis.get_label()

xlab.set_style('italic')
xlab.set_size(15)
ylab.set_style('italic')
ylab.set_size(15)

# tweak the title
ttl = ax.title
ttl.set_weight('bold')
ttl.set_size(20)

plt.plot(years_new,india_new,'--',c='red',label='Predictions using Linear Regression')
plt.plot(years,india_arr,'*',color='r')
plt.grid(False)
plt.legend()
plt.show()
```
