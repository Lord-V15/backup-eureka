```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import keras
from keras.layers import Dense, Activation
from keras.models import Sequential
from keras.utils import np_utils
```

```python
from sklearn.datasets import make_circles

X,y = make_circles(n_samples=2000,factor=0.4)

plt.scatter(X[:,0],X[:,1],c=y)
plt.show()
```

```python
np.unique(y, return_counts=True) #Total values=1000+1000=2000
```

```python
# 1-hot encoding
Y = np_utils.to_categorical(y)
print(Y.shape)
print(Y[:10])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,Y)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape )
```

```python
model = Sequential()
model.add( Dense(5, input_shape=(2,)) )
model.add(Activation('relu'))

model.add(Dense(2))#OUTPUT
model.add(Activation('softmax'))

model.summary()
```

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy']) #Accuracy because this is a classification. if it was a regression, we'd use MSE
```

```python
model.fit(X_train,y_train,batch_size=16, nb_epoch=20, verbose=2, validation_data=(X_test,y_test))
```
